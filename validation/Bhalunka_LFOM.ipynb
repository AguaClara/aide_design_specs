```
# import core aguaclara pkg functions
from aguaclara.design.lfom import LFOM
from aguaclara.core.units import u

# define an LFOM object to be used for some calculations
lfom = LFOM(q=1*u.L/u.s, hl=10*u.cm)

print("An LFOM with a flow rate of 1 L/s and head loss of 10 cm should have the following parameters\norifices per row: {!s}\norifice diameter: {!s}\nspace between rows: {!s}".format(lfom.orifice_n_per_row, lfom.orifice_d, lfom.row_b))
```
